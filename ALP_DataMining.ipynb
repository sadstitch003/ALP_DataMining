## Data Transformation

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.columns = df.columns
df_scaled.head()
```

## Clustering

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, n_init=5, max_iter=500, random_state=0)
    kmeans.fit(df_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
plt.show()
```

```python
from kneed import KneeLocator

findknee = KneeLocator(range(1, 11),
                       wcss,
                       curve="convex",
                       direction="decreasing")

print('the best cluster for your model is', findknee.elbow)
```

```python
km = KMeans(n_clusters=3)
df["KMeansClusters"] = km.fit_predict(df_scaled)
```

```python
sns.scatterplot(data=df, x='Admission Grade', y='Curricular Units 1st Sem (Enrolled)', hue='KMeansClusters', palette='Set1')
plt.title("Hasil KMeans Clusters")
plt.show()
```

```python
ac = AgglomerativeClustering(n_clusters=3)
df["HierarchicalClusters"] = ac.fit_predict(df_scaled)
```

```python
sns.scatterplot(data=df, x='Admission Grade', y='Curricular Units 1st Sem (Enrolled)', hue='HierarchicalClusters', palette='Set1')
plt.title("Hasil Hierarchical Clusters")
plt.show()
```

## Classification

```python
x = df_scaled.drop(columns = "Target")
y = df["Target"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

### Logistic Regression

```python
classifier = LogisticRegression()
classifier.fit(x_train, y_train)
y_pred_logreg = classifier.predict(x_test)
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
print(f"Logistic Regression Accuracy: {accuracy_logreg:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_logreg)}")
```

### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Accuracy: {accuracy_dt:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_dt)}")
```

### Random Forest

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_rf)}")
```

### XGBoost

```python
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)
print(f"XGBoost Accuracy: {accuracy_xgb:.4f}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, y_pred_xgb)}")
```

```python
results = [accuracy_logreg, accuracy_dt, accuracy_rf, accuracy_xgb]
label = ["Logistic Regression", "Decision Tree", "Random Forest", "XGBoost"]

sns.barplot(x = label, y = results)
plt.title("Classifier Accuracy Score Comparison")
plt.show()
```
